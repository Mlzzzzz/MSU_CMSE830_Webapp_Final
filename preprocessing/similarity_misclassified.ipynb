```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Mlzzzzz/MSU_CMSE830_Webapp/main/Datasets/preprocessed_dataset.csv")
df.head()
```

Define the function to calculate the predicted similarity.

```python
def calc_pred_simi(x1, x2):
    y = -5.605 + 5.214*x1 + 2.009*x2
    p = (np.exp(y))/(1 + np.exp(y))
    return p
```

```python
x1 = df['tanimoto_cdk_Extended']
x2 = df['TanimotoCombo']
```

```python
df['pred_frac_simi'] = calc_pred_simi(x1, x2)
```

```python
df.head()
```

Use a threshold of 0.5 to map pred_frac_similar to a new column "pred_similar".

```python
df['pred_similar'] = df['pred_frac_simi'].apply(lambda x: 1 if x > 0.5 else 0)
```

```python
df.head()
```

Filter out the misclassified data points.

```python
df_misclassify = df[df["similar"] != df["pred_similar"]]
df_misclassify
```

```python
df_misclassify.to_csv("misclassified_dataset.csv", index=False)
```
