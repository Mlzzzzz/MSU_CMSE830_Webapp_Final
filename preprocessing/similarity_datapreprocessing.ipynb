```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Mlzzzzz/MSU_CMSE830_Webapp/main/Datasets/new_dataset.csv")
df.head()
```

Drop columns that contain repeating information.

```python
df = df.drop(["curated_smiles_molecule_a", "curated_smiles_molecule_b", "simil_2D", "simil_3D", "dissimil_2D", "dissimil_3D", "n_answers", "n_similar"], axis=1)
```

Use a threshold of 0.5 to map frac_similar to a new column "similar".

```python
df['similar'] = df['frac_similar'].apply(lambda x: 1 if x > 0.5 else 0)
```
